```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

## Q1: Create two charts using list and dataframe.

```python
a=pd.DataFrame([44,33,44,55],columns=['mohammed'])
plt.plot([5,2,3,4],a,c='red',marker='o');
plt.plot([2,5,4,3],[44,33,44,55],marker='*');

```

## Q2: Create pie chart with equal explode values.

```python
plt.figure(figsize=[10,10])
plt.subplot(1,2,1)
plt.pie([1,1,1],labels=['eat','sleep','travel']);

plt.subplot(1,2,2)
plt.pie([22,22,22],labels=['mo','kha','hus']);
```

## Q3: Create a scatter chart with circle marker, alpha = 0.4, and size = 50.

```python
plt.scatter(["hi","hello","goodmorning",'hoy'],[1,2,3,7],marker='o',alpha=0.4,s=50);
```

## Q4: Add the following to your chart:
- title placed in the left and has a red color
- x and y label with font size = 16

```python
plt.scatter(["hi","hello","goodmorning",'hoy'],[1,2,3,7],marker='o',alpha=0.4,s=50);
plt.title('titel',c='red',loc='left');
plt.xlabel('welcoming words',fontsize =16)
plt.ylabel('count of use',fontsize =16)
```

## Q5: Create three line charts with Legends at left and colors.

```python
plt.plot(['mo','so','do'],[3,3,3],c='b');
plt.plot(['mo','so','do'],[2,2,2],c='g');
plt.plot(['mo','so','do'],[1,1,1],c='y');
plt.yticks(range(0,6));
plt.legend(['mm','ss','dd'],loc= 'upper left');
```

## Q6: Create subplots with three charts and figsize=20,10.

```python
plt.figure(figsize=[20,10])

plt.subplot(2,2,1);
plt.plot(['mo','so','do'],[3,3,3],c='b');
plt.legend(['mm'],loc= 'upper left');
plt.yticks(range(0,6));

plt.subplot(2,2,2);
plt.plot(['mo','so','do'],[2,2,2],c='g');
plt.legend(['ss'],loc= 'upper left');
plt.yticks(range(0,6));

plt.subplot(2,2,3);
plt.plot(['mo','so','do'],[1,1,1],c='y');
plt.legend(['dd'],loc= 'upper left');
plt.yticks(range(0,6));


```
